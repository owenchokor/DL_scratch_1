## Activation Function
- **step function**: function that changes its output according to a threshold
- In neural networks, various activation functions are used as the step function

#### step function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    y = x>0
    return y.astype(int)
def sigmoid (x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5, 5, 0.1)
y1 = step_function(x)
y2 = sigmoid(x)
plt.plot(x, y1, color = 'red')
plt.plot(x, y2, color = 'blue')
plt.show()
```
